```python
import numpy as np
import sklearn.datasets
import sklearn.decomposition.pca as PCA
import pandas as pd
from sklearn.preprocessing import StandardScaler

from matplotlib import pyplot as plt
```

```python
#help(PCA)
```

```python
boston = sklearn.datasets.load_boston()
```

```python
boston.data.shape
```

```python
boston_scaled = StandardScaler().fit_transform(boston.data)
#boston_scaled = boston.data
pca = PCA.PCA()
```

```python
boston_pca = pca.fit_transform(boston_scaled)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
pca = PCA.PCA(n_components=2)
boston_pca = pca.fit_transform(boston_scaled)
print(pca.explained_variance_ratio_)
```

2 components explains ~58 percent of our variance.

```python
boston_df = pd.DataFrame(data=boston_pca, columns=['pc1','pc2'])
```

```python
plt.scatter(x=boston_df.pc1, y=boston_df.pc2)
```

Let's add target to the plot

```python
#help(pd.cut)
```

```python
boston_df['target_cat'] = pd.cut(boston.target, 10, labels=range(10))
```

```python
plt.scatter(x=boston_df.pc1, y=boston_df.pc2, c=boston_df.target_cat)
```
